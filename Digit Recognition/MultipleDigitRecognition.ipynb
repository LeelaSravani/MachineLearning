```python
features = dataset.iloc[:,1:]
labels = dataset.iloc[:,0]
```

```python
#labels.head()
features.fillna(features.mean(),inplace = True)
```

```python
from sklearn import preprocessing
features = preprocessing.scale(features)
```

```python
list_hog_fd = []
for feature in features:
    fd = hog(feature.reshape((28, 28)), orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualise=False)
    list_hog_fd.append(fd)
hog_features = np.array(list_hog_fd, 'float64')
```

```python
mlp = MLPClassifier()
```

```python
mlp.fit(hog_features,labels)
```

```python
joblib.dump(mlp, "digits_cls.pkl", compress=3)
```

```python
import cv2
from sklearn.externals import joblib
from skimage.feature import hog
import numpy as np

# Load the classifier
clf = joblib.load("digits_cls.pkl")

# Read the input image 
im = cv2.imread(r"C:\Users\ADMIN\Desktop\PD II\HandWrittenDigits\number.jpg")
cv2.imshow("Input Image image",im)
# Convert to grayscale and apply Gaussian filtering
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray image",im_gray)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
im_gray = cv2.GaussianBlur(im_gray, (5, 5), 0)
cv2.imshow("Gaussian Blur",im_gray)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python

#cv2.imshow("blur",im_gray)
# Threshold the image
ret, im_th = cv2.threshold(im_gray,220, 255,cv2.THRESH_BINARY_INV)
cv2.imshow("Thresholded",im_th)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

_,ctrs, hier = cv2.findContours(im_th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Get rectangles contains each contour
rects = [cv2.boundingRect(ctr) for ctr in ctrs]

# For each rectangular region, calculate HOG features and predict
for rect in rects:
    # Draw the rectangles
    cv2.rectangle(im, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 3) 
    # Make the rectangular region around the digit
    leng = int(rect[3] * 1.6)
    pt1 = int(rect[1] + rect[3] // 2 - leng // 2)
    pt2 = int(rect[0] + rect[2] // 2 - leng // 2)
    roi = im_th[pt1:pt1+leng, pt2:pt2+leng]
    # Resize the image
    roi = cv2.resize(roi, (28, 28), interpolation=cv2.INTER_AREA)
    roi = cv2.dilate(roi, (3, 3))
    # Calculate the HOG features
    roi_hog_fd = hog(roi, orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualise=False)
    nbr = clf.predict(np.array([roi_hog_fd], 'float64')) 
    
    cv2.putText(im, str(int(nbr[0])), (rect[0], rect[1]),cv2.FONT_HERSHEY_DUPLEX, 2, (255, 0, 0), 3)
#print("Predicted digit " ,nbr)
cv2.imshow("Resulting Image with Rectangules drawn around", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
